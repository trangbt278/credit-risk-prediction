```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

# Reprocessing Train Data (2019 loan)

```python
train_df.head()
```

```python
#remove loan_status column
X_train = train_df.drop('loan_status', axis=1)
X_train.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_train = pd.get_dummies(X_train)
X_train.head()
```

```python
# Converting output labels to 0 and 1
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
y_train = LabelEncoder().fit_transform(train_df['loan_status'])
y_train
```

# Reprocessing Test Data (2020 loan)

```python
test_df.head()
```

```python
#remove loan_status column
X_test = test_df.drop('loan_status', axis=1)
X_test.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test = pd.get_dummies(X_test)
X_test.head()
```

```python
# add missing dummy variables to testing set
for i in list(X_train):
    if i not in list(X_test):
        X_test[i] = 0
        
X_test.head()
```

```python
# Converting output labels to 0 and 1
y_test = LabelEncoder().fit_transform(test_df['loan_status'])
y_test
```

```python
from sklearn.metrics import classification_report
target_names = ["high_risk", "low_risk"]
def test_model(model, data):
    X_train, X_test, y_train, y_test = data
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Training Data Score: {model.score(X_train, y_train)}")
    print(f"Testing Data Score: {model.score(X_test, y_test)}")
    
```

```python
data = [X_train, X_test, y_train, y_test]
```

# K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
test_model(KNeighborsRegressor(), data)
```

# Logistic Regression Model

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
test_model(LogisticRegression(), data)
```

# Random Forest Classifier Model

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
test_model(RandomForestClassifier(), data)
```

# Result Assessment

In the logistic regression model, the gap between R^2 on training data and test data  is small so the model fit to full data set does not have many noisy parameters to confuse the model and we get an accurate model.

In the random forest classifier model and the KNearest model, the gap between R^2 on training data and test data  is bigger so the model fit to full data set is overfitting data and the noisy parameters to confuse the model.


# Scale Data

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
data = [X_train_scaled, X_test_scaled, y_train, y_test]
```

# K Nearest Neighbors with scaled data

```python
test_model(KNeighborsRegressor(), data)
```

# Logistic Regression Model with scaled data

```python
# Train a Logistic Regression model on the scaled data and print the model score
test_model(LogisticRegression(), data)
```

# Random Forest Classifier model on the scaled data

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
test_model(RandomForestClassifier(), data)
```

# Result Assessment

In the logistic regression model, the gap between R^2 on scaled training data and test data  is smaller than on the fullset data and we get more accurate model with a small set of data.

In the random forest classifier model and the KNearest model, the gap between R^2 on training data and test data  is still big so there are still noisy parameters confuse the model.

# Undersampling for Imbalanced data 

```python
#undersampling data with NearMiss version 1
from imblearn.under_sampling import NearMiss
undersample = NearMiss(version=1, n_neighbors=3)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with NearMiss version 2
undersample = NearMiss(version=2, n_neighbors=3)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with NearMiss version 3
undersample = NearMiss(version=3, n_neighbors=3)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with CondensedNearestNeighbour
from imblearn.under_sampling import CondensedNearestNeighbour
undersample = CondensedNearestNeighbour(n_neighbors=1)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with TomekLinks
from imblearn.under_sampling import TomekLinks
undersample = TomekLinks()
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with EditedNearestNeighbours
from imblearn.under_sampling import EditedNearestNeighbours
undersample = EditedNearestNeighbours(n_neighbors=3)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with OneSidedSelection
from imblearn.under_sampling import OneSidedSelection
undersample = OneSidedSelection(n_neighbors=1, n_seeds_S=200)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

```python
#undersampling data with NeighbourhoodCleaningRule
from imblearn.under_sampling import NeighbourhoodCleaningRule
undersample = NeighbourhoodCleaningRule(n_neighbors=3, threshold_cleaning=0.5)
undersample = OneSidedSelection(n_neighbors=1, n_seeds_S=200)
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

# Oversampling data

```python
#oversampling data with SMOTE
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_undersample, y_undersample = oversample.fit_resample(X_train, y_train)
data = [X_undersample, X_test, y_undersample, y_test]
test_model(RandomForestClassifier(), data)
```

# Assessment

Apply undersampling and oversampling for imbalanced training data and use RandomForestClassifier model to train the new data set and predict on the test data but the result looks similar, the model is still overfitting.
